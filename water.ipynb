```python
import pandas as pd
import altair as alt
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv("data/uw-supplier-data060722 3.csv",na_values="NA")

df['reporting_month'] = pd.to_datetime(df['reporting_month'], format = '%Y-%m-%d')
```

```python
pd.options.display.max_columns = None
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#How many and what type of suppliers were there in total?
df.supplier_name.value_counts()
```

```python
#What are the different hydrologic regions in the area? How many suppliers are in each region?
df.hydrologic_region.value_counts()
```

```python
#What is the total statewide water production?
df.groupby(by='reporting_month').calculated_total_potable_water_production_gallons_ag_excluded.sum()
```

```python
pd.DataFrame(df.groupby(by='reporting_month').calculated_total_potable_water_production_gallons_ag_excluded.sum()).reset_index()
```

```python
statewide_production=pd.DataFrame(df.groupby(by='reporting_month').calculated_total_potable_water_production_gallons_ag_excluded.sum()).reset_index()
```

```python
alt.Chart(statewide_production).mark_bar(color='#5ba3cf',opacity=0.8).encode(
    y='year(reporting_month):T',
    x='mean(calculated_total_potable_water_production_gallons_ag_excluded):Q',
    tooltip=['mean(calculated_total_potable_water_production_gallons_ag_excluded):Q']
)
```

```python
statewide_production.to_csv("statewide_production.csv", index=False)
```

```python
#What is the total statewide water consumption?
df.groupby(by='reporting_month').calculated_r_gpcd.sum()
```

```python
pd.DataFrame(df.groupby(by='reporting_month').calculated_r_gpcd.sum()).reset_index()
```

```python
statewide_usage=pd.DataFrame(df.groupby(by='reporting_month').calculated_r_gpcd.sum()).reset_index()
```

```python
alt.Chart(statewide_usage).mark_line().encode(x='year(reporting_month):T',y='mean(calculated_r_gpcd):Q')
```

```python
statewide_usage.to_csv("statewide_usage.csv", index=False)
```

```python
#What were the highest and lowest water production months? 
alt.Chart(statewide_production, title='Water production tops in July and August'
).mark_bar(color='#5ba3cf',opacity=0.8).encode(
    x=alt.X('month(reporting_month):T'),
    y='mean(calculated_total_potable_water_production_gallons_ag_excluded):Q',
    tooltip=['mean(calculated_total_potable_water_production_gallons_ag_excluded):Q']
    )
```

```python
#How much water does each hydrolic region produce? Which region has the highest water production?
df.groupby(['hydrologic_region', 'reporting_month']).calculated_total_potable_water_production_gallons_ag_excluded.sum()
```

```python
pd.DataFrame(df.groupby(['hydrologic_region', 'reporting_month']).calculated_total_potable_water_production_gallons_ag_excluded.sum()).reset_index()
```

```python
total_production=pd.DataFrame(df.groupby(['hydrologic_region', 'reporting_month']).calculated_total_potable_water_production_gallons_ag_excluded.sum()).reset_index()
```

```python
alt.Chart(total_production, title='Which region produces the most water?').mark_area(line={'color':'#3887c0'},
    color=alt.Gradient(
        gradient='linear',
        stops=[alt.GradientStop(color='#9cc8e2', offset=0),
               alt.GradientStop(color='#3887c0', offset=1)],
        x1=1,
        x2=1,
        y1=1,
        y2=0
    )).encode(
    tooltip=['reporting_month:T', 'calculated_total_potable_water_production_gallons_ag_excluded:Q'],
    x='yearmonth(reporting_month):T',
    y='calculated_total_potable_water_production_gallons_ag_excluded:Q').facet(facet='hydrologic_region:N',columns=2)
```

```python
#How much water does a resident use per day and how does it varies across hyrdrologic regions?  How does it change from 2014-2022?

df.groupby(['hydrologic_region', 'reporting_month']).calculated_r_gpcd.sum()
```

```python
pd.DataFrame(df.groupby(['hydrologic_region', 'reporting_month']).calculated_r_gpcd.sum()).reset_index()
```

```python
total_usage=pd.DataFrame(df.groupby(['hydrologic_region', 'reporting_month']).calculated_r_gpcd.sum()).reset_index()
```

```python
alt.Chart(total_usage).mark_line().encode(x='reporting_month:T',y='calculated_r_gpcd:Q').facet(facet='hydrologic_region:N',columns=2)
```
